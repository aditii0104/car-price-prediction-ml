```python
from sklearn.preprocessing import LabelEncoder

columns = ['model', 'transmission', 'fuelType']
xlable = df.copy()
label_encoders = {}

for col in columns:
    xlable[col] = xlable[col].astype(str).str.strip()  # clean text
    le = LabelEncoder()
    xlable[col] = le.fit_transform(xlable[col])        # will be 0 if only 1 category
    label_encoders[col] = le

xlable
```
